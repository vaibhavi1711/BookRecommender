```python
from langchain.schema import Document
import os

# Function to safely load the file
def safe_load_file(file_path):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File '{file_path}' does not exist. Please check the file path.")
    if not os.access(file_path, os.R_OK):
        raise PermissionError(f"File '{file_path}' is not readable. Check file permissions.")
    try:
        loader = TextLoader(file_path, encoding="utf-8")  # Specify encoding to handle non-ASCII content
        return loader.load()
    except Exception as e:
        raise RuntimeError(f"Error while loading file '{file_path}': {e}")


# Main program
file_path = "tagged_description.txt"  # Modify this to the absolute path if necessary
try:
    raw_documents = safe_load_file(file_path)
    text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0,
                                          separator="\n")  # Adjust chunk_size and chunk_overlap as needed
    documents = text_splitter.split_documents(raw_documents)
    print("File successfully loaded and split into documents.")
except Exception as e:
    print(f"Error: {e}")
```

```python
documents[0]
```

```python
db_books = Chroma.from_documents(
    documents,
    embedding=OpenAIEmbeddings()
)
```

```python
query = " A book to teach children about nature"
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k = 50)

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]
    return books[books["isbn13"].isin(books_list)].head(top_k)
```

```python
retrieve_semantic_recommendations("A book to teach children about nature")
```
